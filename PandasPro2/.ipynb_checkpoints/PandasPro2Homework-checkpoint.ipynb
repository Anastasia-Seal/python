```python
Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для 
данного значения userId.
```

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
mn = ratings.groupby('userId').max()['timestamp']
mn.head()
```

```python
ratings.groupby('userId').min()['timestamp']
mx.head()
```

```python
df = ratings.groupby('userId')['timestamp'].nunique().to_frame()
raw = df[df['timestamp'] > 100]
raw.rename(columns={'timestamp': 'marks_amount'}, inplace=True)
raw1 = raw.join(mx)

raw1.rename(columns={'timestamp': 'minimum'}, inplace=True)
total = raw1.join(mn)
total.rename(columns={'timestamp': 'maximum'}, inplace=True)

total['result'] = total['maximum'] - total['minimum']

total.head(10)
```

```python
itog = total['result'].sum() / total['result'].count()
itog
```

```python
Задание 2

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). 
Необходимо сформировать две таблицы:
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
аналогичную таблицу по типам выручки с указанием адреса клиента
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:

    rzd - железнодорожные перевозки
    auto - автомобильные перевозки
    air - воздушные перевозки
    client_base - адреса клиентов
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
fr1 = rzd.merge(auto, how = 'outer', on = 'client_id')
fr2 = fr1.merge(air, how = 'outer', on = 'client_id')
fr2
```

```python
fr2.merge(client_base,how ='left', on = 'client_id')
```
